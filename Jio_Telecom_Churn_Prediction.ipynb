### Creating new Columns for predicting Churn

```python
# Finding out total recharge amount of data by multiplying  'av_rech_amt_data_9' and 'total_rech_data_9' In 'total_amt_data_rech_9'
# Derived one Column from total amount paid cutomer in september month by adding 'total_amt_data_rech_9' & total_rech_amt_9
# Derived one Column from total data usges by adding vol_2g_mb_9 & vol_3g_mb_9

df['total_amt_data_rech_9'] = df['total_rech_data_9'] * df['av_rech_amt_data_9']
df['total_amount_paid_9'] = df['total_rech_amt_9'] + df['total_amt_data_rech_9']
df['data_vol_9'] = df['vol_2g_mb_9'] + df['vol_3g_mb_9']

df.head()
```

### Churn Prediction

```python
# Adding Churn Column and giving it value 0[for not churn] or 1[for churn] with the help of following Columns

# arpu_9
# total_ic_mou_9
# total_amount_paid_9
# data_vol_9

# Note- Now i am using only 2 columns from above list, finding out the way to use all these columns in next version of notebook.
# We are comparing arpu_9 with the value 7 because it is giving in the problem statement of the project.
# Not so many customers are using data_vol_9 so we are comparing this value with 0.
# Many Customers use JIO services only for incoming calls so we are comparing it with 0 minutes.
# total_amount_paid_9 coulmn also have nan values and minimum value of this column is 9 
# We are converting it's Nan values to zero and using value 5 for the comparison.

df['total_amount_paid_9'].fillna(0, inplace=True)
 
df["churn"] = np.where(
    (df['arpu_9'] <7) & (df['total_amount_paid_9'] <5) ,
   1, 
   0,)
# Note- Still not able to add & (df['data_vol_9'] <0) & (df['total_ic_mou_9'] <0) these conditions on above syntex, 
# plz provide solution

# df["churn"]=[1500 if x =='Music' else 800 for x in df['Event']]
df.head()
```

### PCA & Decision Tree

```python
# implementing the decision tree on PCA data
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import r2_score
X_train, X_test, y_train, y_test = train_test_split(df_x_pca, y, test_size = 0.3, random_state = 0)
```

```python
max_depth_range = list(range(1, 35))

accuracy = []
for depth in max_depth_range:
    
    clf = DecisionTreeClassifier(max_depth = depth,random_state = 0)
    clf.fit(X_train, y_train)   
    score = clf.score(X_test, y_test)
    accuracy.append(score)
    
     
print(accuracy)
```

```python

dtro = DecisionTreeClassifier(criterion = "gini", #mean sqaure error
                                  random_state = 0,
                                  max_depth=32, 
                                 )
dtro.fit(X_train, y_train)
```

```python
# making predictions with optimal parameters
y_pred = dtro.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
```

```python
# hyperparameter tunning
#  gamma how many support vectors to choose 

# if g is low  more support vector (all points are considred for drawing hyperplne)  high accuracy
# if g is high  few support vectors are considred (only nearly ponits choosen)       low accuracy
 


model=SVC(kernel='rbf',random_state=0,gamma=7)
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```

```python
model=SVC(kernel='rbf',random_state=0,gamma=21)
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```

```python
# C  cost parameter/  cost for misclassification
# if c is low  priority to have maximum marginal distance
# if c is high priority is to classify well.  overfitting arrises as penality is high for mis classification
# by c parameter we are saying our svm how much misclassification to avoid 

model=SVC(kernel='rbf',random_state=0,C=1)
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```

```python
model=SVC(kernel='rbf',random_state=0,C=2)
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```
